```python
import numpy as np
import matplotlib.pyplot as plt

def norm(x, P):
    return np.sqrt(x.T @ P @ x)


P = np.array([[2, 1], [1, 2]])  # Define the matrix P (positive definite)
c = np.array([1, 1])
x = np.linspace(-2, 4, 100)
y = np.linspace(-2, 4, 100)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

norms = np.apply_along_axis(lambda x: norm(x - c, P), 0, points)
inside_set = norms <= 1
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(points[0, inside_set], points[1, inside_set], color='blue', label=r'S')
plt.scatter(c[0], c[1], color='red', marker='o', label='Center c')
plt.title('Convex Set S')
plt.legend()
plt.grid(True)
plt.show()
```
