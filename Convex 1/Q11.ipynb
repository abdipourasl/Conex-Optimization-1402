```python
import numpy as np
# !pip install cvxpy
import cvxpy as cp
import matplotlib.pyplot as plt
import matplotlib
```

```python
x1 = cp.Variable(1)
x2 = cp.Variable(1)
objective = cp.Minimize(cp.power(x1, 2) + cp.multiply(9, cp.power(x2, 2)))
constraints = [(3 * x1) + x2 >= 1, x1 + (2 * x2) >= 1, x1>=0, x2>=0]
prob = cp.Problem(objective, constraints)
print('The optimal value for [{}] is: {}'.format('x1^2 + 9 * x2^2', round(prob.solve(qcp=True), 5)))
print('The optimal x1, x2 is: ({}, {})'.format(x1.value, x2.value))
```

```python
plt.figure
xp = np.array([-2, 20])
plt.plot(xp, (1 - xp) / 3, xp, 1 - (2 * xp))

path = matplotlib.path.Path([[0.4, 0.2], [0, 1], [1, 1], [1, 0]])
patch = matplotlib.patches.PathPatch(path)
plt.gca().add_patch(patch)
plt.xlabel('x2')
plt.ylabel('x1')
plt.title('Feasible Set')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.text(0.7, 0.15, "3 * x1 + x2 = 1")
plt.text(0.15, 0.8, "x1 + 2 * x2 = 1")
plt.scatter(x2.value, x1.value, color='red', marker='o', label='Center c')
plt.text(x2.value, x1.value, "Optimal Point")
plt.show()
```

```python
x1 = cp.Variable(1)
x2 = cp.Variable(1)
objective = cp.Minimize(x1 + 3*x2)
constraints = [-x1 + x2 <= 2, x1 + x2 >= 2, x2>=0, 2*x1 -3*x2 <=5]
prob = cp.Problem(objective, constraints)
print('The optimal value for [{}] is: {}'.format('x1 + 3*x2', round(prob.solve(qcp=True), 5)))
print('The optimal x1, x2 is: ({}, {})'.format(x1.value, x2.value))
```

```python
plt.figure
xp = np.array([-2, 20])
plt.plot(xp, (-2 + xp), xp, (2 - xp) , xp, (5 + (3 * xp))/2, [[0] for _ in xp], xp)
plt.plot()

path = matplotlib.path.Path([[0, 2.5], [0, 2], [2, 0], [5,2.5], [5, 5], [5/3,5]])
patch = matplotlib.patches.PathPatch(path)
plt.gca().add_patch(patch)
plt.xlabel('x2')
plt.ylabel('x1')
plt.title('Feasible Set')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.text(2.7, 1.5, "-x1 + x2 = 2")
plt.text(2.5, -0.5, "x1 + x2 = 2")
plt.text(1.5, 4, "2 * x1 - 3 * x2 = 5")
plt.text(-0.6, 4.5, "x2 = 0")
plt.scatter(x2.value, x1.value, color='red', marker='o', label='Center c')
plt.text(x2.value, x1.value, "Optimal Point")

plt.show()
```
